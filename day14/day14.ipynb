```python
%cd ~/advent_of_code_2024
from utils.get_input import get_input
data = get_input(14).strip()
data = data.splitlines()
import re
data = [re.findall("-*\d+", row) for row in data]
data = [[int(x) for x in row] for row in data]
data = [[row[:2],row[2:]] for row in data]
```

```python
w, h = 101, 103

def move_robots(seconds, w=101, h=103):
    robot_pos = {}
    
    for robot in data:
        x,y = robot[0]
        dx,dy = robot[1]
        new_pos = ((x + dx*seconds) % w, (y + dy*seconds) % h)
        if new_pos in robot_pos:
            robot_pos[new_pos] += 1
        else:
            robot_pos[new_pos] = 1
    return robot_pos
```

```python
robot_pos = move_robots(100)

quads = [0,0,0,0]
for r,n in robot_pos.items():
    x,y = r
    if x < w//2 and y < h//2:
        quads[0] += n
    elif x > w//2 and y < h//2:
        quads[1] += n
    elif x < w//2 and y > h//2:
        quads[2] += n
    elif x > w//2 and y > h//2:
        quads[3] += n

quads[0]*quads[1]*quads[2]*quads[3]
```

```python
from functools import cmp_to_key

def print_map():
    for y in range(h):
        line = ""
        for x in range(w):
            if (x,y) in robot_pos:
                line += str(robot_pos[(x,y)])
            else:
                line += '.'
        print(line)

def d_btwn_p(p1,p2):
    return (((p1[0]-p2[0])**2)+((p1[1]-p2[1])**2))**.5
    
def compare(robot1, robot2):
    return d_btwn_p(robot1,(0,0)) - d_btwn_p(robot2,(0,0))

smallest_average_distance = 100
t = 0
for s in range(10000):
    robot_pos = move_robots(s)
    robots = list(robot_pos.keys())
    robots = sorted(robots, key=cmp_to_key(compare))
    
    distances = [d_btwn_p(r1,r2) for r1,r2 in zip(robots, robots[1:])]
    average_distance = sum(distances) / len(distances)
    
    if average_distance < smallest_average_distance:
        smallest_average_distance = average_distance
        t = s

robot_pos = move_robots(t)
print(t)
print_map()
```
