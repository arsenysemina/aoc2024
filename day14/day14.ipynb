```python
%cd ~/advent_of_code_2024
from utils.get_input import get_input
data = get_input(14).strip()
data = data.splitlines()
import re
data = [re.findall("-*\d+", row) for row in data]
data = [[int(x) for x in row] for row in data]
data = [[row[:2],row[2:]] for row in data]
```

```python
w, h = 101, 103
def move_robots(seconds, w=101, h=103):
    robot_pos = {}
    
    for robot in data:
        x,y = robot[0]
        dx,dy = robot[1]
        new_pos = ((x + dx*seconds) % w, (y + dy*seconds) % h)
        if new_pos in robot_pos:
            robot_pos[new_pos] += 1
        else:
            robot_pos[new_pos] = 1
    return robot_pos
```

```python
from functools import reduce
from operator import mul

robot_pos = move_robots(100)
divider = ((w//2),(h//2))

quads = [0,0,0,0]
for r,n in robot_pos.items():
    x,y = r
    if x < divider[0] and y < divider[1]:
        quads[0] += n
    elif x > divider[0] and y < divider[1]:
        quads[1] += n
    elif x < divider[0] and y > divider[1]:
        quads[2] += n
    elif x > divider[0] and y > divider[1]:
        quads[3] += n

reduce(mul, quads)
```
